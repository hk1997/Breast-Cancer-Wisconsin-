First we are going to start by visualising data and then apply various models with different hyperparameter tunings to see their effect.


```python
#importing necessary dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#reading dataset
dataset = pd.read_csv('./data.csv')
```

```python
dataset.head()
```

```python
dataset.describe()
```

We find that as of now, id and Unnamed: 32 features are not useful and diagnosis is the label therefore we need to remove id and Unnamed: 32 from our dataset.

```python
y=dataset.diagnosis
list=['id','Unnamed: 32','diagnosis']
x=dataset.drop(list,axis=1)
```

```python
#Now visualising the distribution of labels in our dataset
v=y.value_counts()
_x=[]
_x.append(v['B'])
_x.append(v['M'])
labels=['Benign','Malignant']
plt.pie(_x,labels=labels)
plt.show()
print(v)
```

```python
#describing columns in our dataset
col=dataset.columns
print(col)
```

Now we are going to start our process of visualizing our data. <br>
Before proceeding, we are going to normalize data to reduce difference between different feature values so they can be visualized effectively. 

```python
x = (x - x.mean())/(x.std())
```

To startwith, we are going to use violinplot to see distribution of values for benign and malignant tumor for each of the features.

```python
import seaborn as sb
def violin_plot(i,j):
    data = pd.concat([y,x.iloc[:,i:j]],axis=1)
    data = pd.melt(data,id_vars="diagnosis",
                    var_name="features",
                    value_name='value')
    plt.figure(figsize=(10,10))
    sb.violinplot(x="features", y="value", hue="diagnosis", data=data,split=True, inner="quart")
    plt.xticks(rotation=90)
    plt.show()
```

Above violin plot give us probabilistic disrtibution of values of features corresponding to M & B.
Several features show similar distribution and hence may be less useful than others showing clear distinction. 

Now we are going to see data distribution using swarm plots


```python

def swarm_plot(i,j):
    data = pd.concat([y,x.iloc[:,i:j]],axis=1)
    data = pd.melt(data,id_vars="diagnosis",
                    var_name="features",
                    value_name='value')
    plt.figure(figsize=(10,10))
    #tic = time.time()
    sb.swarmplot(x="features", y="value", hue="diagnosis", data=data)
    plt.xticks(rotation=90)
    plt.show()
```

```python
swarm_plot(0,8)
```

```python
swarm_plot(8,16)
```

```python
swarm_plot(16,24)
```

```python
swarm_plot(24,32)
```

Through above swarm plots we can see variance more clearly. Also we are able to figure out which classifiers which features look more clearer in terms of classification. 

```python
plt.figure(figsize=(18,18))
sb.heatmap(x.corr(),annot=True,linewidth=0.5,fmt='.1f')
plt.show()
```

Above heatmap gives us correlation between all pairs of features. <br>
We can reduce unwanted features by taking only one feature of pairs having high degree of correlation.

/*-before removing some features let's run random forest classifier over all the features and notice accuracy--*/

```python
#loading dependencies for randomforest classifier
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score,confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
# split data train 70 % and test 30 %
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
#random forest classifier with n_estimators=10 (default)
clf_rf = RandomForestClassifier(random_state=43)      
print('Now fitting data')
clr_rf = clf_rf.fit(x_train,y_train)
print('data fitting completed')
```

```python
ac = accuracy_score(y_test,clf_rf.predict(x_test))
print('Accuracy is: ',ac)
cm = confusion_matrix(y_test,clf_rf.predict(x_test))
sb.heatmap(cm,annot=True,fmt="d")
plt.show()
```

Now we are going to remove features having high correlation coffecient b/w them
<br>
they are:
1)<B> Compactness_mean, concavity_mean and concave points_mean</B>
<br>
2) <B>radius_se, perimeter_se and area_se</B>
<br>
3) <B> radius_worst, perimeter_worst and area_worst </B>
<br>
4) <B> Compactness_worst, concavity_worst and concave points_worst </B>
<br>
5) <B> Compactness_se, concavity_se and concave points_se  </B>
<br>
6) <B> texture_mean and texture_worst <B>
<br>
7) <B> texture_mean. area_worst and area_mean </B

Now we are going to choose any one of the features in the above list out of correlated features, hence we will make our drop list


```python
drop_list1 = ['perimeter_mean','radius_mean','compactness_mean','concave points_mean','radius_se','perimeter_se',
              'radius_worst','perimeter_worst','compactness_worst','concave points_worst','compactness_se',
              'concave points_se','texture_worst','area_worst']
x_1 = x.drop(drop_list1,axis = 1) #dropping these features
x_1.head()
```

```python
#creating a heatmap to see degree of correlation 
plt.figure(figsize=(18,18))
sb.heatmap(x_1.corr(), annot=True, linewidth=0.5, fmt='0.1f')
plt.show()
```

As we can see, degree of correlation have decreased. <br>
Now we are going to apply our randomforest classifier again on the dataset.

```python
# split data train 70 % and test 30 %
x_train, x_test, y_train, y_test = train_test_split(x_1, y, test_size=0.3, random_state=42)
#random forest classifier with n_estimators=10 (default)
clf_rf = RandomForestClassifier(random_state=43)      
print('Now fitting data')
clr_rf = clf_rf.fit(x_train,y_train)
print('data fitting completed')
```

```python
ac = accuracy_score(y_test,clf_rf.predict(x_test))
print('Accuracy is: ',ac)
cm = confusion_matrix(y_test,clf_rf.predict(x_test))
sb.heatmap(cm,annot=True,fmt="d")
plt.show()
```

Accuracy before removing correlated features: Accuracy is:  0.947368421053
<br>
Thus, accuracy has been increased
